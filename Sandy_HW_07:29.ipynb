```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import re
```

#### HW1:

1. read file
2. count the number of unique words in the file
3. counte the occurrence frequeny of each word in the file
4. get the most popular word, least popular word
5. separate words with characters other than alphabets, get the count for each group
6. standardize your text 
7. visualize the dataset: occurrence frequency of top 100 words ( line plot, bar graph, scatter plot)

```python
words = []
with open('wikipedia.txt', 'r') as fp:
    for line in fp:
        line = line.strip()
        if line:
            words.extend([x.lower() for x in line.split()])
unique_words = set(words)
unique_word_count = len(unique_words)
print('Unique words count: ', unique_word_count)
print(unique_words)
```

```python
word_freq = Counter(words)
most_popular_word = word_freq.most_common(1)[0]
least_popular_word = word_freq.most_common()[-1]
print("Most and least popular words are:", most_popular_word[0], "and", least_popular_word[0])
```

```python
non_alpha_groups = [x for x in words if not x.isalpha()] 
print(non_alpha_groups)
non_alpha_groups_count = Counter(non_alpha_groups)
```

```python
standardized_words = [re.sub(r'[^\w\s]', '', x) for x in words]
standardized_freq = Counter(standardized_words)
standardized_words
```

```python
top_words = standardized_freq.most_common(100)
print(top_words)
words_top, counts_top = zip(*top_words)
```

```python
plt.figure(figsize=(15, 5))
plt.bar(words_top, counts_top)
plt.title('Bar Graph: Top 100 Words by Frequency')
plt.xticks(rotation=90)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

#### HW2:

Read the csv wine_quality.csv into a DataFrame and perform the following operations:

1. Create a new dataframe with an equal number of columns as the wine_quality.csv. Each column of this new dataframe should contain boolean values.The value should be true if the corresponding column from wine_quality is an outlier else it should be False. Use Interquartile range to calculate outliers. (Lower bound: Q1-1.5* IQR Upper bound:Q2+1.5*IQR IQR = Q3-Q1)

2. For each column of this new data frame, get count of outliers and normal values.

3. Filter out all the outliers from the dataframe

4. Infuse NaNs into your original dataframe

5. Create a function to perform linear interpolation to approximate the NaN values. Compare the interpolated values with the values as calculated by the predefined interpolation method provided by pandas.

6. Plot line chart for all the columns of the wine_quality.csv

```python
df = pd.read_csv('wine_quality.csv')

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df < lower_bound) | (df > upper_bound))
outliers

outlier_counts = outliers.sum()
normal_counts = (~outliers).sum()
print("Outlier counts per column:\n", outlier_counts)
print("\nNormal counts per column:\n", normal_counts)
```

```python
df_no_outliers = df[~outliers.any(axis=1)]
df_no_outliers
```

```python
nan_df = df.mask(np.random.random(df.shape) < 0.1)
nan_df
```

```python
def self_interpolate(column):
    data = column.copy()
    n = len(data)
    for i in range(n):
        if pd.isna(data[i]):
            prev_index = i - 1
            next_index = i + 1
            while prev_index >= 0 and pd.isna(data[prev_index]):
                prev_index -= 1
            while next_index < n and pd.isna(data[next_index]):
                next_index += 1
                
            if prev_index >= 0 and next_index < n:
                y1, y2 = data[prev_index], data[next_index]
                x1, x2 = prev_index, next_index
                data[i] = y1 + ((y2 - y1) / (x2 - x1)) * (i - x1)
    return data

interpolated_df_custom = nan_df.apply(self_interpolate)
interpolated_df_custom
    
```

```python
interpolated_df_pandas = nan_df.interpolate()
interpolated_df_pandas
```

```python
for column in df.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(df[column], label='Original')
    plt.plot(interpolated_df_pandas[column], label='Pandas Interpolated', linestyle='--')
    plt.plot(interpolated_df_custom[column], label='Custom Interpolated', linestyle='-.')
    plt.title(f'Line Chart for {column}')
    plt.legend()
    plt.show()
```
